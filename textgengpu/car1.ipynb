#####  训练模型

```python
textgen = textgenrnn(name="car1")                   # 给模型起个名字，比如`my.poem`, 之后生成的模型文件都会以这个名字为前缀
textgen.reset()                                         # 重置模型
with tf.device('/device:GPU:0'):
    textgen.train_from_file(                                # 从数据文件训练模型
        file_path = '/home/aily/shitao/ben_c_trim.txt'  ,          # 文件路径
        new_model = True,                                   # 训练新模型
        num_epochs = 20,                                    # 训练轮数
#         word_level = True,                                 # True:词级别，False:字级别
#         rnn_bidirectional = True,                           # 是否使用Bi-LSTM
#         max_length = 100                                  # 一条数据的最大长度 
        train_size=0.8,
        dropout=0.3,
        
    )
```

##### 生成数据
