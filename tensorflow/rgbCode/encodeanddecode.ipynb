### RGB 编码解码


```python
import matplotlib.pyplot as plt
import tensorflow as tf

#读取原始图片
image = tf.gfile.FastGFile("/home/aily/桌面/pan.jpeg", 'rb').read()
with tf.Session() as sess:
    #解码
    img_after_decode = tf.image.decode_jpeg(image)
    
    #获取解码矩阵,得到可视化矩阵
    print(img_after_decode.eval())
    plt.imshow(img_after_decode.eval())
    plt.show()
    encode_image = tf.image.encode_jpeg(img_after_decode)
    with tf.gfile.GFile("home/aily/桌面/pan.jpeg", 'wb') as f:
        f.write(encode_image.eval())
```
